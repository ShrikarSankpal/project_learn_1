```python
import pandas as pd
```

# Putting NA for missing values

```python
ccr = pd.read_csv('risk_factors_cervical_cancer.csv',na_values='?')
```

```python
ccr.head()
```

# Drop Columns where NAs are 90%

```python
ccr = ccr.loc[:,ccr.columns[ccr.isnull().sum() / ccr.shape[0] <= 0.9 ]]
```

```python
ccr.count()
```

```python
ccr.head()
```

# Drop columns which are having strings or objects

```python
ccr.dtypes
```

```python
ccr = ccr.select_dtypes(exclude=['object'])   #No object dtype here
```

```python
ccr.head()
```

```python
for col in ccr.columns:
    print(col, ccr[col].unique())
```

### converting columns where unique values are 3(0,1,NaN) to bool

```python
for col in ccr.columns:
    if len(ccr[col].unique())<=3:
        ccr[col]=ccr[col].astype('bool')
```

```python
ccr.dtypes
```

```python
for a in ccr.columns:
    print(ccr[a].dtypes)
```

## Print colmuns that have minimum one NaN value

```python
#df.columns[df.isna().any()]
ccr.isna().any()
```

```python
ccr.describe()
```

# Fill missing values Int or float values by median and bool by mode

```python
for col in ccr.columns[ccr.isnull().any()]:
    if ccr[col].dtype != 'bool':
        ccr[col].fillna(ccr[col].median(),inplace=True)
    else:
        ccr[col].fillna(ccr[col].mode(),inplace=True)
```

```python
ccr.head()
```

```python
ccr.describe()
```

```python
ccr.describe(include = 'all')
```

```python
ccr.columns[ccr.isnull().any()]
```

```python
ccr.head()
```

# Merging all the target variables into 1

```python
ccr['target'] = ccr.iloc[:,-1] | ccr.iloc[:,-2] | ccr.iloc[:,-3] | ccr.iloc[:,-4]
```

```python
ccr['target'].unique()
```

```python
ccr.head()
```

## Removing the 4 target variables (Hinselmann	Schiller	Citology	Biopsy)

```python
ccr.drop(ccr.iloc[:,-5:-1],axis =1,inplace=True)
```

```python
ccr.head()
```

## checking for all rows having same values

```python
(ccr.iloc[0] == ccr).all()
```

# Normalizing data using Min Max scaler

```python
from sklearn import preprocessing
```

```python
x = ccr.values #returns a numpy array
x
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
min_max_scaler
```

```python
x_scaled = min_max_scaler.fit_transform(x)
x_scaled
```

```python
ccr = pd.DataFrame(x_scaled,columns=ccr.columns)
```

```python
ccr.describe(include= 'all')
```

```python
# Save the new cleaned data
ccr.to_csv("cleaned_riskfactor_cervical_cancer.csv",index=False)
```
